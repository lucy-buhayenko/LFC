```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
```

# Problem 1

```python
def r_generated_boxes(image, n):
    #dimensions of image array
    h, w, color = image.shape
    
    boxes = np.zeros((n,4), dtype=int)

    for i in range(n):
        r1 = np.random.randint(0, h-1)
        r2 = np.random.randint(r1 + 1, h)

        
        c1 = np.random.randint(0, w-1)
        c2 = np.random.randint(c1 + 1, w)

        boxes[i] = [r1, c1, r2, c2]

    return boxes

        
```

```python
image = io.imread('carnegie.jpg')
n = 5
boxes = r_generated_boxes(image, n)
```

```python
plt.imshow(image)
for box in boxes:
    r1, c1, r2, c2 = box
    
    corners = np.array([[c1, r1], #top left
                        [c2, r1], #top right
                        [c2, r2], #bottom right
                        [c1, r2], #bottom left
                        [c1, r1]])#top left again
    plt.plot(corners[:,0], corners[:,1], color='yellow', linewidth=2)

plt.axis('off')
plt.show()
```

# Problem 2

```python
def red_rectangle(image, r1, c1, r2, c2):
    # top edge
    image[r1, c1:c2+1] = [255, 0, 0]
    #bottom edge
    image[r2, c1:c2+1] = [255, 0, 0]
    #left edge
    image[r1:r2+1, c1] = [255, 0, 0]
    # right edge
    image[r1:r2+1, c2] = [255, 0, 0]

    return image
```

```python
image = io.imread('LFC.jpg')
r1,c1,r2,c2 = 50, 100, 150, 250
red_rectangle(image,r1,c1,r2,c2)
plt.imshow(image)
plt.axis('off')
plt.show()
```
